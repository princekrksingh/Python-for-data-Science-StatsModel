# real_estate_price_size_year Regression

```python
#impoting libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
#load the data
data = pd.read_csv("/home/prince/Desktop/real_estate_price_size_year.csv")
```

```python
data
```

```python
#declare variables
y = data['price']
x1 = data[['size','year']]
```

```python
#regression
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
#Addition of year variable increses R-squared as well Adj r-squared value that means 
#variable has significant effect on model.
```
